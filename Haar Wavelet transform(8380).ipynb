### K means

```python
from sklearn.cluster import KMeans
```

```python
X_array_sd1.shape
```

### Elbow method

```python
cluster_range = range( 1, 20 )
cluster_errors = []

for num_clusters in cluster_range:
    clusters = KMeans( num_clusters )
    clusters.fit( X_array_sd1 )
    cluster_errors.append( clusters.inertia_ )
    
clusters_df = pd.DataFrame( { "num_clusters":cluster_range, "cluster_errors": cluster_errors } )
plt.figure(figsize=(12,6))
plt.plot( clusters_df.num_clusters, clusters_df.cluster_errors, marker = "o" )
```

### Choosing number of clusters as 3

```python
# Number of clusters
kmeans = KMeans(n_clusters=3)
# Fitting the input data
kmeans = kmeans.fit(X_array_sd1)
# Getting the cluster labels
labels = kmeans.predict(X_array_sd1)
# Centroid values
centroids = kmeans.cluster_centers_

#Unique labels
np.unique(labels)
```

```python
zero=0
one=0
two=0
for i in labels:
    if i==0:
        zero=zero+1
    elif i==1:
        one=one+1
    elif i==2:
        two=two+1
```

```python
print("Number of entries with 0 as the label :" , zero)
print("Number of entries with 1 as the label :" , one)
print("Number of entries with 2 as the label :" , two)
```

```python
X_array_sd1_df=pd.DataFrame(X_array_sd1)
X_array_sd1_df['Labels']=labels
X_array_sd1_df.columns = ["PC1","PC2","PC3","Labels"]
```

### Three dimensional plot for visualisation

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.figure(figsize=(100,100))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = X_array_sd1_df["PC1"]
Y = X_array_sd1_df["PC2"]
Z = X_array_sd1_df["PC3"]

ax.scatter(X,Y,Z, c=X_array_sd1_df["Labels"], marker ='o')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
```
